```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape')
#to avoid encoding error, use unicode_escape
```

```python
df.dropna(inplace=True)
```

```python
#change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
#rename column
df.rename(columns={'Marital_Status':'Shaadi'},inplace='True')
```

EDA

Gender

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)    #adds title to the bars
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x= 'Gender',y= 'Amount', data=sales_gen)
```

AGE

```python
az = sns.countplot(data =df,x='Age Group', hue='Gender')

for new in az.containers:
    az.bar_label(new)
```

```python
df.groupby(['Age Group'], as_index=True)['Amount'].sum().sort_values(ascending=False)
```

```python
sale_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = "Amount",ascending=False)
sns.barplot(x= 'Age Group',y= 'Amount', data=sale_age)
```

STATE

```python
sale_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = "Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15, 8)})
sns.barplot(x= 'State',y= 'Orders', data=sale_state)
```

```python
sale_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = "Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15, 8)})
sns.barplot(x= 'State',y= 'Amount', data=sale_states)
```

Marital Status

```python
ka = sns.countplot(x='Marital_Status',data=df)

for mar in ka.containers:
    ka.bar_label(mar)
    
sns.set(rc={'figure.figsize':(7,5)})
```

```python
sale_married = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by = "Amount",ascending=False)
sns.barplot(x="Marital_Status", y="Amount", data=sale_married, hue='Gender')
```

OCCUPATION

```python
a=sns.countplot(x='Occupation',data=df)

for y in a.containers:
    a.bar_label(y)

sns.set(rc={'figure.figsize':(19,7)})
```

```python
sale_occupation = df.groupby(['Occupation'],as_index=False)["Amount"].sum().sort_values(by="Occupation")
sns.barplot(x='Occupation',y='Amount',data=sale_occupation)
```

Product Category

```python
c = sns.countplot(x='Product_Category',data=df)

sns.set(rc={'figure.figsize':(25,7)})
for h in c.containers:
    c.bar_label(h)
```

```python
sale_pc = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Product_Category',y='Amount',data=sale_pc)
```
